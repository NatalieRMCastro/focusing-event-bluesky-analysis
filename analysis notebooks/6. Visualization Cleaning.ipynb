## 2. Visualizations

### 2.1 BlueSky Posts - Climate Focusing Windows

```python
## Adding a new line character here to get the bar to wrap
climate_focusing = [{'Dictionary':'Environmental Action & \nEnvironmental Justice',"Total Frequency":11539},
                      {'Dictionary':'Policy',"Total Frequency":8350},
                     {'Dictionary':'Climate Change',"Total Frequency":6568},
                     {'Dictionary':'Republican',"Total Frequency":5168},
                     {'Dictionary':'Democrat',"Total Frequency":4393},
                     {'Dictionary':'Science',"Total Frequency":3735},
                     {'Dictionary':"Focusing Events",'Total Frequency':2853}]
values = pd.DataFrame(climate_focusing)
```

```python
values.head()
```

```python
sb.set_style("white")
sb.set(font='Times New Roman', font_scale=1.2)
fig, ax = plt.subplots(figsize=(12,5 )) 



hist = sb.barplot(values,x="Dictionary",y="Total Frequency",color="black")


for bar_group in hist.containers:  # `containers` contains the histogram bars grouped by hue
    # Get heights and centers for each group
    heights = [bar.get_height() for bar in bar_group]
    centers = [bar.get_x() + bar.get_width() / 2 for bar in bar_group]
    
    # Add markers
    plt.scatter(centers, heights, color='black', zorder=3)
    
    for center, height in zip(centers, heights):
        plt.text(
            center, 
            height, 
            f'{int(height):,}',  # Format with commas
            color='black', 
            ha='center', 
            va='bottom', 
            fontsize=10
        )

for container in hist.containers:
    hist.bar_label(container, fmt='%d', label_type='edge', padding=3, bbox=dict(boxstyle="round,pad=0.3", edgecolor="black", facecolor="white"))

hist.set_title("BlueSky Posts By Dictionary Category: Climate Focusing Events")
hist.set_xlabel("Dictionary Type")
hist.set_ylabel("Posts")
ax.tick_params(axis='x')
ax.grid(True)
ax.set_facecolor("white")

formatter = FuncFormatter(lambda x, _: f'{int(x):,}')
ax.yaxis.set_major_formatter(formatter)


# Set the border (box) around the entire plot
for spine in ax.spines.values():
    spine.set_edgecolor('black')  # Set the color of the border to black
    spine.set_linewidth(1)  # Set the width of the border

plt.tight_layout()
plt.savefig("Distribution of BlueSky Posts By Dictionary Category - Climate Focusing Events.png",dpi=2500)
plt.show();
```

```python
## Generating alt text
alt_text = f"Bar chart showing {len(values['Dictionary'])} categories: "
alt_text += ", ".join([f"{cat} ({val})" for cat, val in zip(values['Dictionary'], values['Total Frequency'])])
alt_text += ". The highest value is " + max(zip(values['Total Frequency'], values['Dictionary']))[1] + " at " + str(max(values['Total Frequency'])) + "."
alt_text += " The lowest value is " + min(zip(values['Total Frequency'], values['Dictionary']))[1] + " at " + str(min(values['Total Frequency'])) + "."

# Print Alt Text
print(alt_text)
```
